# 1 - Aprendizaje supervisado: Regresión por CM

Vamos a implementar el algoritmo de regresión por cuadrados mínimos.

```python
# Refactor for least squares, and then we can visualize.
def least_sq_loss(X, y, theta):
    return .5 * np.sum((X @ theta - y)**2)

def render_points(X, y, points, delta=0.2, isocline=False, loss=least_sq_loss):
    (theta_0,theta_1) = (np.arange(-5,5,delta), np.arange(-5,5,delta))
    J = loss(X, theta_0, theta_1, y, delta=delta)
    T0,T1 = np.meshgrid(theta_0,theta_1)
    if isocline:
        fig, ax = plt.subplots()
        ax.set_title('Gradient Descent')
        CS = ax.contour(T0, T1, J)
    else:
        fig = plt.figure()
        ax = plt.axes(projection='3d')
        ax.set_title('Gradient Descent')
        CS = ax.contour3D(T0, T1, J, 50)
    ax.clabel(CS, inline=1, fontsize=10)
    def animate(i): ax.plot( points[i][0], points[i][1], "or")
    ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(points), repeat=False)
    return ani
```

```python
X_train = np.loadtxt('data_01/x.dat')
y_train = np.loadtxt('data_01/y.dat')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_train[:,0], X_train[:,1], y_train, c="r", marker="o")
plt.show()
```

```python
X_train, y_train = map(torch.Tensor, (X_train, y_train))
```

```python
class LinearRegression(nn.Module):
    def __init__(self, in_features, out_features=1):
        super(LinearRegression, self).__init__()
        
        # Defino el objeto Linear que hace de layer, es
        # llamable y tiene guardado los parámetros.
        self.Linear = nn.Linear(in_features,out_features)
        
    def forward(self,x):
        return self.Linear(x)       
    
N, d = X_train.shape
model = LinearRegression(d)
out = model(X_train)
```
